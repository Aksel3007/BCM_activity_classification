```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hs_csv = pd.read_csv("BCM-activity-classification(2).csv")

# List the keys in the csv file
print(hs_csv.keys())

# Separate into 5 different dataframes
ceps = hs_csv[hs_csv['metadata/file_path'] == "data/bcm/"] # cepstral
spec = hs_csv[hs_csv['metadata/file_path'] == "data/bcm_spectrograms/"] # spectrogram
```

```python
sns.set_theme(style="darkgrid", palette="deep")
# seaborn styles are: white, dark, whitegrid, darkgrid, ticks


# Plot the boxplot
# Set the figure size
plt.figure(figsize=(10, 8))
sns.set(font_scale=2.2)

# Create the boxplot
sns.boxplot([spec['training/val_acc_epoch (max)']*100, ceps['training/val_acc_epoch (max)']*100], width = 0.4)
sns.despine(left=True)
plt.xticks([0,1],['Mel-spectrogram', 'MFCC'])
# Add title and axis names
plt.title('Spectrogram vs MFCC',fontweight="bold")
plt.ylabel('Validation Accuracy (%)',fontweight="bold")
plt.xlabel('Preprocessing',fontweight="bold")

plt.savefig('Spec_vs_ceps.png', dpi=300,bbox_inches='tight')

plt.show()
```
