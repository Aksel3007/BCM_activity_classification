```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set_theme(style="dark", palette="deep")
# seaborn styles are: white, dark, whitegrid, darkgrid, ticks

'''
import the following files:

full_datasets_data.csv
indiv_dataset_seed.csv
indiv_datasets_data.csv
loso_dataset_seed.csv
loso_datasets_data.csv
full_dataset_seed.csv
finetune_dataset_data.csv
'''

full_datasets_data = pd.read_csv("full_datasets_data.csv")
indiv_dataset_seed = pd.read_csv("indiv_dataset_seed.csv")
indiv_datasets_data = pd.read_csv("indiv_datasets_data.csv")
loso_dataset_seed = pd.read_csv("loso_dataset_seed.csv")
loso_datasets_data = pd.read_csv("loso_datasets_data.csv")
full_dataset_seed = pd.read_csv("full_dataset_seed.csv")
finetune_dataset_data = pd.read_csv("finetune_dataset_data.csv")

# For each file, find the standard deviation of the "Val acc (last)" column
full_datasets_data_std = full_datasets_data["Val acc (last)"].std()
indiv_dataset_seed_std = indiv_dataset_seed["Val acc (last)"].std()
indiv_datasets_data_std = indiv_datasets_data["Val acc (last)"].std()
loso_dataset_seed_std = loso_dataset_seed["Val acc (last)"].std()
loso_datasets_data_std = loso_datasets_data["Val acc (last)"].std()
full_dataset_seed_std = full_dataset_seed["Val acc (last)"].std()
finetune_dataset_data_std = finetune_dataset_data["Val acc (last)"].std()

# Print all the standart deviations
print("full_datasets_data_std: ", full_datasets_data_std)
print("indiv_dataset_seed_std: ", indiv_dataset_seed_std)
print("indiv_datasets_data_std: ", indiv_datasets_data_std)
print("loso_dataset_seed_std: ", loso_dataset_seed_std)
print("loso_datasets_data_std: ", loso_datasets_data_std)
print("full_dataset_seed_std: ", full_dataset_seed_std)
print("finetune_dataset_data_std: ", finetune_dataset_data_std)

```

```python
spacing = 0.06

print(full_datasets_data.shape[0])
x_full = np.zeros(full_datasets_data.shape[0])
x_indiv = np.zeros(indiv_datasets_data.shape[0])+0.6
x_loso = np.zeros(loso_datasets_data.shape[0])+1.2+spacing
x_finetune = np.zeros(finetune_dataset_data.shape[0])+1.8

# Calculate the mean of all validation accuracies
full_datasets_data_mean = full_datasets_data["Val acc (last)"].mean()
indiv_datasets_data_mean = indiv_datasets_data["Val acc (last)"].mean()
loso_datasets_data_mean = loso_datasets_data["Val acc (last)"].mean()
finetune_dataset_data_mean = finetune_dataset_data["Val acc (last)"].mean()


# Plot the validation accuracy as scatterplot for each file, all on the same plot, with the dataset type on the x-axis
plt.figure(figsize=(20, 10))

# Add horizontal lines for the mean of the each dataset, and limit the y-axis
plt.hlines(y=100*full_datasets_data_mean, xmin=0, xmax=spacing*2, colors="grey", linestyles="dashed")
plt.hlines(y=100*indiv_datasets_data_mean, xmin=0.6, xmax=0.6+spacing*2, colors="grey", linestyles="dashed")
plt.hlines(y=100*loso_datasets_data_mean, xmin=1.2, xmax=1.2+spacing*2, colors="grey", linestyles="dashed")
plt.hlines(y=100*finetune_dataset_data_mean, xmin=1.8, xmax=1.8+spacing*2, colors="grey", linestyles="dashed")

# Print all the means
print("full_datasets_data_mean: ", full_datasets_data_mean)
print("indiv_datasets_data_mean: ", indiv_datasets_data_mean)
print("loso_datasets_data_mean: ", loso_datasets_data_mean)
print("finetune_dataset_data_mean: ", finetune_dataset_data_mean)


# Add different markers for each dataset type
# Seaborn marker codes: o, v, ^, <, >, 8, s, p, P, *, h, H, +, x, X, D, d, |, _, 1, 2, 3, 4
# seaborn marker size syntax: markersize=number
marker_size = 200
# Marker colors in sns scatterplot syntax is: color=string
# Default color is blue, but you can also use: "r", "g", "b", "c", "m", "y", "k", "w"
# Split the full_dataset_data into 3 equal parts, and plot each part on the same plot
sns.scatterplot(y=100*full_datasets_data["Val acc (last)"][0:full_datasets_data.shape[0]//3],x = x_full[0:full_datasets_data.shape[0]//3]+2*spacing, label="full_datasets_data", marker="P",color = 'c', s = marker_size)
sns.scatterplot(y=100*full_datasets_data["Val acc (last)"][full_datasets_data.shape[0]//3:2*full_datasets_data.shape[0]//3],x = x_full[full_datasets_data.shape[0]//3:2*full_datasets_data.shape[0]//3]+spacing, label="full_datasets_data", marker=">", color = 'm', s = marker_size)
sns.scatterplot(y=100*full_datasets_data["Val acc (last)"][2*full_datasets_data.shape[0]//3:],x = x_full[2*full_datasets_data.shape[0]//3:], label="full_datasets_data", color = 'y', s = marker_size)

# Split the indiv_dataset_data into 3 equal parts, and plot each part on the same plot
sns.scatterplot(y=100*indiv_datasets_data["Val acc (last)"][0:indiv_datasets_data.shape[0]//3],x = x_indiv[0:indiv_datasets_data.shape[0]//3]+2*spacing, label="indiv_datasets_data", marker="P",color = 'c', s = marker_size)
sns.scatterplot(y=100*indiv_datasets_data["Val acc (last)"][indiv_datasets_data.shape[0]//3:2*indiv_datasets_data.shape[0]//3],x = x_indiv[indiv_datasets_data.shape[0]//3:2*indiv_datasets_data.shape[0]//3]+spacing, label="indiv_datasets_data", marker=">", color = 'm', s = marker_size)
sns.scatterplot(y=100*indiv_datasets_data["Val acc (last)"][2*indiv_datasets_data.shape[0]//3:],x = x_indiv[2*indiv_datasets_data.shape[0]//3:], label="indiv_datasets_data", color = 'y', s = marker_size)

# Loso
sns.scatterplot(y=100*loso_datasets_data["Val acc (last)"][0],x = x_loso+spacing, label="loso_datasets_data",  marker="P",color = 'c', s = marker_size)
sns.scatterplot(y=100*loso_datasets_data["Val acc (last)"][1],x = x_loso, label="loso_datasets_data", marker=">", color = 'm', s = marker_size)
sns.scatterplot(y=100*loso_datasets_data["Val acc (last)"][2],x = x_loso-spacing, label="loso_datasets_data", color = 'y', s = marker_size)

# Finetune, split into 3 equal parts, and plot each part on the same plot
sns.scatterplot(y=100*finetune_dataset_data["Val acc (last)"][0:finetune_dataset_data.shape[0]//3],x = x_finetune[0:finetune_dataset_data.shape[0]//3], label="finetune_dataset_data", color = 'y', s = marker_size)
sns.scatterplot(y=100*finetune_dataset_data["Val acc (last)"][finetune_dataset_data.shape[0]//3:2*finetune_dataset_data.shape[0]//3],x = x_finetune[finetune_dataset_data.shape[0]//3:2*finetune_dataset_data.shape[0]//3]+spacing, label="finetune_dataset_data", marker=">", color = 'm', s = marker_size)
sns.scatterplot(y=100*finetune_dataset_data["Val acc (last)"][2*finetune_dataset_data.shape[0]//3:],x = x_finetune[2*finetune_dataset_data.shape[0]//3:]+2*spacing, label="finetune_dataset_data", marker="P",color = 'c', s = marker_size)

# No legend
plt.legend().remove()



# Set xticks, but with no vertical lines

plt.xticks([0, 0.6, 1.2, 1.8], ["Full Datasets", "Individual Datasets", "LOSO Datasets", "Fine-tuning Datasets"])

# Font size seaborn
sns.set(font_scale=3)


# Add legend describing that dashed lines are the mean of the dataset
plt.legend(["Mean Accuracy","Subject 1","Subject 2","Subject 3"], loc="best")



plt.xlabel("Dataset Type",fontweight="bold") # Bold font
plt.ylabel("Validation Accuracy (%)", fontweight="bold")
plt.title("Validation Accuracy For Each Dataset Type",fontweight="bold")

# save the result as high dpi image, with tight layout
plt.tight_layout()
plt.savefig("val_acc_all.png", dpi=200, bbox_inches='tight')

plt.show()
```
