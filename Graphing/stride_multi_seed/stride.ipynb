```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hs_csv = pd.read_csv("BCM-activity-classification(2).csv")

# List the keys in the csv file
print(hs_csv.keys())

# Separate into 5 different dataframes
s1 = hs_csv[hs_csv['metadata/stride (average)'] == 0.032] # stride = 0.032
s2 = hs_csv[hs_csv['metadata/stride (average)'] == 0.1] # stride = 0.1
s3 = hs_csv[hs_csv['metadata/stride (average)'] == 0.5] # stride = 0.5
s4 = hs_csv[hs_csv['metadata/stride (average)'] == 1] # stride = 1
s5 = hs_csv[hs_csv['metadata/stride (average)'] == 3] # stride = 3
```

```python
sns.set_theme(style="darkgrid", palette="deep")
# seaborn styles are: white, dark, whitegrid, darkgrid, ticks

# Plot the boxplot
# Set the figure size
plt.figure(figsize=(10, 8))
sns.set(font_scale=2.2)

# Create the boxplot
sns.boxplot([s1['training/val_acc_epoch (max)']*100, s2['training/val_acc_epoch (max)']*100, s3['training/val_acc_epoch (max)']*100, s4['training/val_acc_epoch (max)']*100, s5['training/val_acc_epoch (max)']*100], width=0.3)
sns.despine(left=True)

# Add title and axis names
plt.title('Stride vs Validation Accuracy',fontweight="bold")
plt.ylabel('Validation Accuracy (%)',fontweight="bold")
plt.xlabel('Stride (Seconds)',fontweight="bold")

# xticks
plt.xticks([0,1, 2, 3, 4], ['0.032', '0.1', '0.5', '1', '3'])

plt.savefig('stride.png', dpi=300,bbox_inches='tight')

plt.show()

```
