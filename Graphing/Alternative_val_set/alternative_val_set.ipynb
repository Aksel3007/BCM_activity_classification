```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hs_csv = pd.read_csv("BCM-activity-classification.csv")

# List the keys in the csv file
print(hs_csv.keys())

# Separate into 5 different dataframes
set1 = hs_csv[hs_csv['metadata/file_path'] == "data/bcm/"] 
set2 = hs_csv[hs_csv['metadata/file_path'] == "data/bcm_alt/"]
set3 = hs_csv[hs_csv['metadata/file_path'] == "data/bcm_alt_2/"]
set4 = hs_csv[hs_csv['metadata/file_path'] == "data/bcm_alt_3/"]


# Plot the boxplot
# Set the figure size
plt.figure(figsize=(10, 7))

# Create the boxplot
#plt.boxplot([spec['training/val_acc_epoch (max)']*100, ceps['training/val_acc_epoch (max)']*100], labels=['Mel-spectrogram', 'MFCC'], widths=0.6)
plt.boxplot([set1['training/val_acc_epoch (max)']*100, set2['training/val_acc_epoch (max)']*100, set3['training/val_acc_epoch (max)']*100, set4['training/val_acc_epoch (max)']*100], labels=['Set 1', 'Set 2', 'Set 3', 'Set 4'], widths=0.6)

# Add title and axis names
plt.title('Comparing the validation accuracy of different validation datasets')
plt.ylabel('Validation Accuracy (%)')
plt.xlabel('Dataset')

plt.show()
```
