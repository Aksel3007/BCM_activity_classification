```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
'''
import the following files:

noise1.csv
noise2.csv
pitch_1_tone.csv
pitch_half_semitone_rerun_new.csv
speed_10_percent.csv
temp_cutout_1s.csv
temp_cutout_200_ms.csv
data_aug_control.csv
aggregate.csv
'''

# import data
noise1 = pd.read_csv('noise1.csv')
noise2 = pd.read_csv('noise2.csv')
pitch_1_tone = pd.read_csv('pitch_1_tone.csv')
pitch_half_semitone_rerun_new = pd.read_csv('pitch_half_semitone_rerun_new.csv')
speed_10_percent = pd.read_csv('speed_10_percent.csv')
speed_20_percent = pd.read_csv('speed2.csv')
temp_cutout_1s = pd.read_csv('temp_cutout_1s.csv')
temp_cutout_200_ms = pd.read_csv('temp_cutout_200_ms.csv')
data_aug_control = pd.read_csv('data_aug_control.csv')
aggregate = pd.read_csv('aggregate_all.csv')

# Make lists for 'Val acc (last)' for each data augmentation method
noise1_list_max = noise1['Val acc (last)'].tolist()
noise2_list_max = noise2['Val acc (last)'].tolist()
pitch_1_tone_list_max = pitch_1_tone['Val acc (last)'].tolist()
pitch_half_semitone_rerun_new_list_max = pitch_half_semitone_rerun_new['Val acc (last)'].tolist()
speed_10_percent_list_max = speed_10_percent['Val acc (last)'].tolist()
speed_20_percent_list_max = speed_20_percent['Val acc (last)'].tolist()
temp_cutout_1s_list_max = temp_cutout_1s['Val acc (last)'].tolist()
temp_cutout_200_ms_list_max = temp_cutout_200_ms['Val acc (last)'].tolist()
data_aug_control_list_max = data_aug_control['Val acc (last)'].tolist()
aggregate_list_max = aggregate['Val acc (last)'].tolist()


# Make lists for 'training/val_acc_epoch (max)' for each data augmentation method
val_acc_max = True

if val_acc_max == True:
    noise1_list_max = noise1['training/val_acc_epoch (max)'].tolist()
    noise2_list_max = noise2['training/val_acc_epoch (max)'].tolist()
    pitch_1_tone_list_max = pitch_1_tone['training/val_acc_epoch (max)'].tolist()
    pitch_half_semitone_rerun_new_list_max = pitch_half_semitone_rerun_new['training/val_acc_epoch (max)'].tolist()
    speed_10_percent_list_max = speed_10_percent['training/val_acc_epoch (max)'].tolist()
    speed_20_percent_list_max = speed_20_percent['training/val_acc_epoch (max)'].tolist()
    temp_cutout_1s_list_max = temp_cutout_1s['training/val_acc_epoch (max)'].tolist()
    temp_cutout_200_ms_list_max = temp_cutout_200_ms['training/val_acc_epoch (max)'].tolist()
    data_aug_control_list_max = data_aug_control['training/val_acc_epoch (max)'].tolist()
    aggregate_list_max = aggregate['training/val_acc_epoch (max)'].tolist()



# Print the mean of each list
print('noise1                           ', np.mean(noise1_list_max))
print('noise2                           ', np.mean(noise2_list_max))
print('pitch_1_tone                     ', np.mean(pitch_1_tone_list_max))
print('pitch_half_semitone_rerun_new    ', np.mean(pitch_half_semitone_rerun_new_list_max))
print('speed_10_percent                 ', np.mean(speed_10_percent_list_max))
print('speed_20_percent                 ', np.mean(speed_20_percent_list_max))
print('temp_cutout_1s                   ', np.mean(temp_cutout_1s_list_max))
print('temp_cutout_200_ms               ', np.mean(temp_cutout_200_ms_list_max))
print('data_aug_control                 ', np.mean(data_aug_control_list_max))
print('aggregate                        ', np.mean(aggregate_list_max))

```
