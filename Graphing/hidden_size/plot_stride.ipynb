```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import the data from the csv files in the folder "stride"
val_acc_83 = pd.read_csv('stride/run-version_83-tag-val_acc_epoch.csv')
val_acc_84 = pd.read_csv('stride/run-version_84-tag-val_acc_epoch.csv')
val_acc_85 = pd.read_csv('stride/run-version_85-tag-val_acc_epoch.csv')
val_acc_86 = pd.read_csv('stride/run-version_86-tag-val_acc_epoch.csv')
val_acc_87 = pd.read_csv('stride/run-version_87-tag-val_acc_epoch.csv')

# Plot the data

fig = plt.figure(figsize=(12, 8))
plt.plot(val_acc_83['Wall time']-val_acc_83['Wall time'][0],val_acc_83['Value']*100, label='Stride: 3 seconds')
plt.plot(val_acc_84['Wall time']-val_acc_84['Wall time'][0],val_acc_84['Value']*100, label='Stride: 2 seconds')
plt.plot(val_acc_85['Wall time']-val_acc_85['Wall time'][0],val_acc_85['Value']*100, label='Stride: 1 seconds')
plt.plot(val_acc_86['Wall time']-val_acc_86['Wall time'][0],val_acc_86['Value']*100, label='Stride: 0.5 seconds')
plt.plot(val_acc_87['Wall time']-val_acc_87['Wall time'][0],val_acc_87['Value']*100, label='Stride: 0.1 seconds')

plt.xlabel('Time [Seconds]')
plt.xlim(0, 1000)
plt.ylabel('Validation Accuracy [%]')
plt.legend()
plt.grid()
fig.patch.set_facecolor('white')
plt.show()


```
