```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import sys

```

```python
# Load the csv files from the folder

run2 = pd.read_csv('hidden_size_train_acc/BCMAC-2__training_train_acc_epoch.csv',names=['epoch','time','train_acc'])
run3 = pd.read_csv('hidden_size_train_acc/BCMAC-3__training_train_acc_epoch.csv',names=['epoch','time','train_acc'])
run4 = pd.read_csv('hidden_size_train_acc/BCMAC-4__training_train_acc_epoch.csv',names=['epoch','time','train_acc'])
run5 = pd.read_csv('hidden_size_train_acc/BCMAC-5__training_train_acc_epoch.csv',names=['epoch','time','train_acc'])
run6 = pd.read_csv('hidden_size_train_acc/BCMAC-6__training_train_acc_epoch.csv',names=['epoch','time','train_acc'])
run14 = pd.read_csv('hidden_size_train_acc/BCMAC-14__training_train_acc_epoch.csv',names=['epoch','time','train_acc'])
run15 = pd.read_csv('hidden_size_train_acc/BCMAC-15__training_train_acc_epoch.csv',names=['epoch','time','train_acc'])

# plot the training accuracy for each run

# Figure size
plt.figure(figsize=(10, 6))

plt.plot(run2['epoch'],run2['train_acc'],label='Hidden size: 64')
plt.plot(run3['epoch'],run3['train_acc'],label='Hidden size: 128')
plt.plot(run4['epoch'],run4['train_acc'],label='Hidden size: 32')
plt.plot(run5['epoch'],run5['train_acc'],label='Hidden size: 8')
plt.plot(run6['epoch'],run6['train_acc'],label='Hidden size: 256')
plt.plot(run14['epoch'],run14['train_acc'],label='Hidden size: 512')
plt.plot(run15['epoch'],run15['train_acc'],label='Hidden size: 1024')

plt.ylim(0.95,1)
plt.xlabel('epoch')
plt.ylabel('train accuracy')
plt.legend()
plt.show()

```
