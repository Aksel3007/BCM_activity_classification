```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft
from scipy import signal
from scipy.io import wavfile
import librosa
import seaborn as sns


```

```python

sns.set_theme(style="darkgrid", palette="deep")
# seaborn styles are: white, dark, whitegrid, darkgrid, ticks
plt.figure(figsize=(10, 6))
sns.set(font_scale=2.2)

n = np.arange(16)
L = 32
lift = 1 + (L/2.)*np.sin(np.pi*n/L)
lift
plt.plot(lift)

# Set x-ticks and y-ticks
plt.xticks(np.arange(0, 16, step=2))
plt.yticks(np.arange(0, 17, step=2))


# Add axis labels and title
plt.xlabel('Cepstral Coefficient',fontweight="bold")
plt.ylabel('Liftering Factor',fontweight="bold")
plt.title('MFCC Liftering',fontweight="bold")
plt.tight_layout()
plt.show()
```
