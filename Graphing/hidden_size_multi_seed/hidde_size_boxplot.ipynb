```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

sns.set_theme(style="darkgrid", palette="deep")
# seaborn styles are: white, dark, whitegrid, darkgrid, ticks


hs_csv = pd.read_csv("BCM-activity-classification.csv")

# List the keys in the csv file
print(hs_csv.keys())

'''plt.plot(hs_csv['metadata/lstm_hidden_size (last)'])
plt.show()

plt.plot(hs_csv['metadata/lstm_hidden_size (last)'], hs_csv['training/val_acc_epoch (max)'])
plt.show()'''

# Separate into 4 different dataframes
hs8 = hs_csv[hs_csv['metadata/lstm_hidden_size (last)'] == 8]
hs64 = hs_csv[hs_csv['metadata/lstm_hidden_size (last)'] == 64]
hs128 = hs_csv[hs_csv['metadata/lstm_hidden_size (last)'] == 128]
hs512 = hs_csv[hs_csv['metadata/lstm_hidden_size (last)'] == 512]

# Plot the boxplot
# Set the figure size
plt.figure(figsize=(10, 8))
sns.set(font_scale=2.2)

sns.boxplot([hs8['training/val_acc_epoch (max)']*100, hs64['training/val_acc_epoch (max)']*100, hs128['training/val_acc_epoch (max)']*100, hs512['training/val_acc_epoch (max)']*100], width=0.3)
sns.despine(left=True)
plt.xticks([0, 1, 2, 3], ['8', '64', '128', '512'])
plt.xlabel('LSTM Hidden Size',fontweight="bold")
plt.ylabel('Validation Accuracy (%)',fontweight="bold")
plt.title('Validation Accuracy vs. LSTM Hidden Size',fontweight="bold")


plt.savefig('output.png', dpi=300,bbox_inches='tight')
plt.show()

```

```python
# Boxplot of training accuracy vs. hidden size
plt.figure(figsize=(10, 6))
plt.boxplot([hs8['training/train_acc_epoch (max)']*100, hs64['training/train_acc_epoch (max)']*100, hs128['training/train_acc_epoch (max)']*100, hs512['training/train_acc_epoch (max)']*100])

plt.xticks([1, 2, 3, 4], ['8', '64', '128', '512'])
plt.xlabel('LSTM Hidden Size')
plt.ylabel('Training Accuracy (%)')
plt.title('Training Accuracy vs. LSTM Hidden Size')

plt.show()
```
