```python
# import the plotting libraies
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 # import the csv files using pd.read_csv:
''' BCMAC-677__training_val_acc_epoch.csv
BCMAC-1355__training_val_acc_epoch.csv
BCMAC-1359__training_val_acc_epoch.csv'''

run677 = pd.read_csv('BCMAC-677__training_val_acc_epoch.csv')
run1355 = pd.read_csv('BCMAC-1355__training_val_acc_epoch.csv')
run1359 = pd.read_csv('BCMAC-1359__training_val_acc_epoch.csv')

sns.set_theme(style="darkgrid", palette="deep")
# seaborn styles are: white, dark, whitegrid, darkgrid, ticks
plt.figure(figsize=(10, 8))
sns.set(font_scale=2.2)


plt.plot(run677.iloc[:,2].rolling(5).mean(), label="run677") # Control
plt.plot(run1355.iloc[:,2].rolling(5).mean(), label="run1355") # Combination of all
plt.plot(run1359.iloc[:,2].rolling(5).mean(), label="run1359") # Combination of 2

# Add title and x, y labels
plt.title("Validation Accuracy", fontweight='bold')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")

# Add legend with the custom labels: COntrol, Combination of all, Combination of 2
plt.legend(labels=['Control', 'Combination of all', 'Combination of 2'])
```
