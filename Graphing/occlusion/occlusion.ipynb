```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
import seaborn as sns
```

```python

sns.set_theme(style="darkgrid", palette="deep")
# seaborn styles are: white, dark, whitegrid, darkgrid, ticks

# Set the figure size
plt.figure(figsize=(10, 8))
sns.set(font_scale=2.2)



# Load the csv files from the folder
run26 = pd.read_csv('occlusion\BCMAC-26__training_val_acc_epoch.csv',names=['epoch','time','train_acc'])
run27 = pd.read_csv('occlusion\BCMAC-27__training_val_acc_epoch.csv',names=['epoch','time','train_acc'])
run16 = pd.read_csv('training_val_acc_epoch.csv',names=['epoch','time','train_acc'])


# Plot the results

plt.plot(run26['epoch'], run26['train_acc']*100, label='Lower Half of Cepstrum Occluded')
plt.plot(run27['epoch'], run27['train_acc']*100, label='Upper Half of Cepstrum Occluded')
plt.plot(run16['epoch'], run16['train_acc']*100, label='No Occlusion')

# Plot title
plt.title('Occlusion Sensitivity (Cepstrum)')

plt.ylim(90,100)
plt.xlim(0,40)
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')
plt.legend()
plt.show()
```

```python
# Load the csv files from the folder

run28 = pd.read_csv('occlusion_spec\BCMAC-28__training_val_acc_epoch.csv',names=['epoch','time','train_acc'])
run29 = pd.read_csv('occlusion_spec\BCMAC-29__training_val_acc_epoch.csv',names=['epoch','time','train_acc']) # Lower frequencies occluded
run30 = pd.read_csv('occlusion_spec\BCMAC-30__training_val_acc_epoch.csv',names=['epoch','time','train_acc']) # Upper frequencies occluded

sns.set_theme(style="darkgrid", palette="deep")
# seaborn styles are: white, dark, whitegrid, darkgrid, ticks

# Set the figure size
plt.figure(figsize=(10, 8))
sns.set(font_scale=2.2)

# Plot the results

plt.figure(figsize=(10, 12))
plt.plot(run28['epoch'], run28['train_acc']*100, label='No occlusion')
plt.plot(run29['epoch'], run29['train_acc']*100, label='Lower Half of Spectrum Occluded')
plt.plot(run30['epoch'], run30['train_acc']*100, label='Upper Half of Spectrum Occluded')


# Plot title
plt.title('Occlusion Sensitivity Spectrograms',fontweight="bold")

plt.xlim(0,200)
#plt.ylim(0.9,1)
plt.xlabel('Epoch',fontweight="bold")
plt.ylabel('Validation Accuracy',fontweight="bold")

# Legend in middle right
plt.legend(loc = 'center right')


plt.show()
```
