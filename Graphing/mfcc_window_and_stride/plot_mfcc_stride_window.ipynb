```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# import data
#mfcc_stride_16ms.csv
#mfcc_window_64ms.csv
#mfcc_window64ms_stride64ms.csv
#data_aug_control.csv

mfcc_stride_16ms = pd.read_csv("mfcc_stride_16ms.csv")
mfcc_window_64ms = pd.read_csv("mfcc_window_64ms.csv")
mfcc_window64ms_stride64ms = pd.read_csv("mfcc_window64ms_stride64ms.csv")
data_aug_control = pd.read_csv("data_aug_control.csv")

# Make lists for 'Val acc (last)' for each data augmentation method
mfcc_stride_16ms_list = mfcc_stride_16ms['Val acc (last)'].tolist()
mfcc_window_64ms_list = mfcc_window_64ms['Val acc (last)'].tolist()
mfcc_window64ms_stride64ms_list = mfcc_window64ms_stride64ms['Val acc (last)'].tolist()
data_aug_control_list = data_aug_control['Val acc (last)'].tolist()

sns.set_theme(style="darkgrid", palette="deep")

plt.figure(figsize=(10, 8))
sns.set(font_scale=2.2)

# Create the boxplot
ax = sns.boxplot(data=[data_aug_control_list, 
                       mfcc_window_64ms_list, 
                       mfcc_stride_16ms_list,
                       mfcc_window64ms_stride64ms_list],
                 width=0.3,  
                 showmeans=True, 
                 meanprops={"marker":"s","markerfacecolor":"white", "markeredgecolor":"black"})


plt.xticks([0 ,1, 2, 3], ['Window: 32 ms\nStride: 32 ms', 'Window: 64 ms\nStride: 32 ms','Window: 32 ms\nStride: 16 ms', 'Window: 64 ms\nStride: 64 ms'])
plt.xticks(rotation=45)

# Add yticks at 0.65, 0.7, 0.75, 0.8, 0.85, 0.9 with the corresponding percentages
plt.yticks(np.arange(0.65, 0.95, 0.05), ['65', '70', '75', '80', '85', '90'])

# Add labels to the plot
plt.xlabel('Window And Stride for MFCCs', fontweight='bold')
plt.ylabel('Validation Accuracy (%)', fontweight='bold')
plt.title('Varying Window and Stride for MFCCs', fontweight='bold')

plt.show()



```
