```python
# Make a list of all the paths to the folders containing the data

def data_paths():
    path_list = []
    for subdir, dirs, files in sorted(os.walk('bcm_behaviour_data_multi_subject')):
        for dir in dirs:
            if not "subject" in dir:
                if "mfcc_window0.064" in dir:
                    print(f'{subdir}/{dir}')
                    path_list.append(f'{subdir}/{dir}')
    return path_list

paths_np = np.array(data_paths())
#np.save('bcm_behaviour_data_multi_subject/paths_mfcc_window0.064.npy', paths_np)
new_list = list(paths_np)
print(new_list)


```

```python
def plot_mfcc(data_mfcc):
    fig = plt.figure(figsize=(30, 5))
    plt.pcolormesh(np.transpose(data_mfcc),shading='auto')
    plt.ylabel('Mel Cepstrum Coefficients')
    plt.xlabel('Time [sec]')
    plt.show()

normal   = np.load('bcm_behaviour_data_multi_subject/subject1/2022-09-20_14-58-39/1.npy')

new_data = np.load('bcm_behaviour_data_multi_subject/subject1/2022-09-20_14-58-39test!/1.npy')

old_data = np.load('bcm_behaviour_data_multi_subject/subject1/2022-09-20_14-58-39mfcc_window0.064/1.npy')


plot_mfcc(normal)
plot_mfcc(new_data)
plot_mfcc(old_data)

for i in range(100):
    print(np.mean(abs(normal[i+100])))
    print(np.mean(abs(new_data[i+100])))
    print(np.mean(abs(old_data[i+100])))
    print('\n')
```
